This Colab Notebook contains the code to preprocess the house loans dataset. Standard Machine Learning Algorithms such as Logistic Regression and SVC are applied to the dataset to predict the Loan Status for a given applications.

```python
import pandas as pd
import numpy as np
```

```python
#The csv file needs to be uploaded to colab
loan_train = pd.read_csv("./loan_sanction_train.csv")
loan_train = loan_train.drop(["Loan_ID"],axis=1)
```

```python
loan_train.head()
```

```python
#transforming the categorical Variables to numerical form
loan_train.Self_Employed = loan_train.Self_Employed.eq('Yes').mul(1)
loan_train.Education = loan_train.Education.eq('Graduate').mul(1)
loan_train.Loan_Status = loan_train.Loan_Status.eq('Y').mul(1)
loan_train.Married = loan_train.Married.eq('Yes').mul(1)
loan_train.Gender = loan_train.Gender.eq('Male').mul(1)

loan_train
```

```python
loan_train.Dependents.unique()
```

```python
loan_train['Dependents'] = loan_train['Dependents'].replace(['3+'], '3')
```

```python

s = (loan_train.dtypes == 'object')
object_cols = list(s[s].index)
object_cols
```

```python
#One hot encoding the "Property Area" column
y = pd.get_dummies(loan_train.Property_Area, prefix='PA')
print(y.head())
loan_train['PA_Rural'] = y['PA_Rural']
loan_train['PA_Semiurban'] = y['PA_Semiurban']
loan_train['PA_Urban'] = y['PA_Urban']

loan_train = loan_train.drop(["Property_Area"],axis=1)
loan_train
```

```python
#dropping entries that have atleast 1 NaN value
loan_train = loan_train.dropna()
loan_train
```

```python
loan_train.Dependents.unique()
```

```python
loan_train_y = loan_train["Loan_Status"]
loan_train = loan_train.drop(["Loan_Status"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(loan_train, loan_train_y, test_size=0.3, random_state=42)
```

```python
#Normalizing the datasets
reqd_cols = ['ApplicantIncome',	'CoapplicantIncome'	,'LoanAmount'	,'Loan_Amount_Term']
for col in reqd_cols:
  initial_mean = X_train[col].mean()
  initial_std = X_train[col].std()

  X_train[col]=(X_train[col]-initial_mean)/initial_std
  X_test[col]=(X_test[col]-initial_mean)/initial_std

X_train.head()
```

```python
#Using the logistic regression model to predict loan applications
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression()

logisticRegr.fit(X_train, y_train)
score = logisticRegr.score(X_test, y_test)

print(f"The test accuracy is {100*score}")
```

```python
#Using the Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf = RandomForestClassifier()
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"The test accuracy is {100*accuracy}")
```

```python
#using SVC classifier

from sklearn.svm import SVC
svc = SVC(gamma='auto')
svc.fit(X_train, y_train)

y_pred = svc.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"The test accuracy is {100*accuracy}")
```
